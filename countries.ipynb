# World boundaries

Load data from [Natural Earth](https://www.naturalearthdata.com/downloads/10m-cultural-vectors/)

```python
import geopandas as gpd

countries_df = gpd.read_file(
    "ne_10m_admin_0_countries.zip",
    engine="pyogrio",
    use_arrow=True,
)
```

Handle notable exceptions

```python
# France
countries_df.loc[countries_df["NAME"] == "France", "ISO_A2"] = "FR"
countries_df.loc[countries_df["NAME"] == "France", "ISO_A3"] = "FRA"
# Norway
countries_df.loc[countries_df["NAME"] == "Norway", "ISO_A2"] = "NO"
countries_df.loc[countries_df["NAME"] == "Norway", "ISO_A3"] = "NOR"
# Kosovo
countries_df.loc[countries_df["NAME"] == "Kosovo", "ISO_A2"] = "XK"
countries_df.loc[countries_df["NAME"] == "Kosovo", "ISO_A3"] = "UNK"
```

Plot countries

```python
countries_df.plot()
```

Keep only interesting columns

```python
countries_df = countries_df[
    ["NAME", "NAME_LONG", "ISO_A2", "ISO_A3", "FORMAL_EN", "geometry"]
]
countries_df = countries_df.rename(
    columns={
        "NAME": "name",
        "NAME_LONG": "name_long",
        "FORMAL_EN": "formal_en",
    }
)
```

Remove countries with no ISO Code

```python
countries_df = countries_df[countries_df["ISO_A2"] != "-99"]
```

Define a function to map each geometry into a `FeatureCollection`

```python
import json
from shapely.geometry import Polygon, MultiPolygon, mapping
from pathlib import Path


def row_to_featurecollection(row) -> dict:
    geom = row.geometry
    properties = row.drop(labels="geometry").to_dict()

    features = []

    if isinstance(geom, MultiPolygon):
        # Split multipolygon into separate polygons
        for poly in geom.geoms:
            features.append(
                {"type": "Feature", "geometry": mapping(poly), "properties": properties}
            )
    elif isinstance(geom, Polygon):
        features.append(
            {"type": "Feature", "geometry": mapping(geom), "properties": properties}
        )
    else:
        # fallback for other geometries (LineString, Point, etc.)
        features.append(
            {"type": "Feature", "geometry": mapping(geom), "properties": properties}
        )

    return {"type": "FeatureCollection", "features": features}
```

Apply the function to all rows to save as ISO A2

```python
for _, row in countries_df.iterrows():
    fc = row_to_featurecollection(row)
    iso_code = row["ISO_A2"]

    # Skip empty / invalid codes
    if not iso_code or iso_code.strip() == "-99":
        continue
    # Save to file
    data_folder = Path("ISO_A2")
    data_folder.mkdir(exist_ok=True)
    filename = data_folder / f"{iso_code}.json"

    with Path.open(filename, "w") as f:
        json.dump(fc, f)
```

Apply the function to all rows to save as ISO A3

```python
for _, row in countries_df.iterrows():
    fc = row_to_featurecollection(row)
    iso_code = row["ISO_A3"]

    # Skip empty / invalid codes
    if not iso_code or iso_code.strip() == "-99":
        continue
    # Save to file
    data_folder = Path("ISO_A3")
    data_folder.mkdir(exist_ok=True)
    filename = data_folder / f"{iso_code}.json"

    with Path.open(filename, "w") as f:
        json.dump(fc, f)
```
